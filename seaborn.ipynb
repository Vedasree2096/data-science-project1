```python
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
df=sb.load_dataset('tips')

print(df.to_string())
g=sb.FacetGrid(df,col="time")
g.map(plt.hist,"tip")
plt.show()
```

```python
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
df=sb.load_dataset('tips')
g=sb.FacetGrid(df,col="sex",hue="smoker")
g.map(plt.scatter,"total_bill","time")
plt.show()
```

```python
#pair plot
#importing packages
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
#loading dataset using seaborns
df=sb.load_dataset('tips')
#df.head()
#print(df.to_string())
sb.pairplot(df)
plt.show()
```

```python
import seaborn as sb
from matplotlib import pyplot as plt
df=sb.load_dataset('tips')
sb.pairplot(df,hue="size", palette="husl")
plt.show()
```

```python
sb.pairplot(df,diag_kind="kde")
plt.show()
```

```python
sb.pairplot(df,kind="reg")
plt.show()
```

```python
sb.pairplot(df,hue="sex",markers=["o","s"])
plt.show()
```

```python
g=sb.pairplot(df,hue="day")
g.fig.suptitle("pairplot wof tips datset",y=1.02)
g.set(xticks=[],yticks=[])
plt.show()
```
